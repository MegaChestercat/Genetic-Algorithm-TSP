## Step 6. Definition of of functions to select parents for next generation

```python
def selection_pair(population: Population, fitness_func:FitnessFunc) -> Population:
  a =  [fitness_func(gene) for gene in population]
  probability = 0.4
  individuals = 0
  pairs = []
  if(sum(a) == 0):
    #return None
    
    
    for i in range(0, len(population)):
      if individuals < 2:
        if random() > probability:
          pairs.append(population[i])
          individuals += 1  

    return pairs
    
    
  else:
    for i in range(0, len(population)):
      if individuals < 2:
        if random() > probability:
          pairs.append(population[i])
          individuals += 1

    return pairs
    
f = selection_pair(population = generate_population(10), fitness_func = partial(fitness, time_limit=4320, lowest_cost = 1500))
print(f)
```

## Step 7. Definition of Crossover function

```python
def single_point_crossover(a: Genome, b: Genome) -> Tuple[Genome, Genome]:
  
  length = len(a)

  if length < 12:  # it only makes sense if genome is at least of length
    return a, b

  p = randint(1, 11)  # find a point where to cut the genomes
  #print(p)
  return a[0:p] + b[p:], b[0:p] + a[p:]

f = selection_pair(population = generate_population(10), fitness_func = partial(fitness, time_limit=4320, lowest_cost = 1500))
print(f)
if f is not None:
  h, j = single_point_crossover(a = f[0], b = f[1])
  print(f"h = {h}")
  print(f"j = {j}")
```

## Step 7.5 Definition of Mutation function

```python
def mutation(genome: Genome, num: int = 1, probability: float = 0.7) -> Genome:
    for _ in range(num):
        index = randrange(len(genome))
        if random() > probability:
          genome[index] = genome[index]
        else:
          genome[index] = index
    return genome

f = selection_pair(population = generate_population(10), fitness_func = partial(fitness, time_limit=4320, lowest_cost = 1500))
print(f)
if f is not None:
  h, j = single_point_crossover(a = f[0], b = f[1])
  print(f"h = {h}")
  print(f"j = {j}")
  k = mutation(h)
  l = mutation(j)
  print(f"k = {k}")
  print(f"l = {l}")
```

## Step 8. Definition of Evolution Function

```python
def run_evolution(
      populate_func: PopulateFunc,
      fitness_func: FitnessFunc,
      fitness_limit: int,
      selection_func: SelectionFunc = selection_pair,
      crossover_func: CrossoverFunc =single_point_crossover,
      mutation_func: MutationFunc = mutation,
      generation_limit: int = 100,
      printer: Optional[PrinterFunc] = None):
    population = populate_func()

    print("")
    print("GENETIC ALGORITHM")
    print("----------")
    print(f"Generation 0")
    print(population)

    for i in range(1, generation_limit+1):
      population = sorted(population, key=lambda genome: fitness_func(genome), reverse=False)
      
      if printer is not None:
            printer(population, i, fitness_func)

      #if fitness_func(population[0]) >= fitness_limit:  # stop if limit is reached or exceeded
           #break
      next_generation = population[0:2] # copy the 2 top solutions

      for j in range(int(len(population) / 2) - 1):
          parents = selection_func(population, fitness_func) # get the parents
          if parents != None:
            offspring_a, offspring_b = crossover_func(parents[0], parents[1])  # generate two child solutions
            offspring_a = mutation_func(offspring_a) # expand the variety of possible solutions by mutating each child
            offspring_b = mutation_func(offspring_b)
            next_generation += [offspring_a, offspring_b]

      population = next_generation  # replace current generation with the next one
      print("")
      print("GENETIC ALGORITHM")
      print("----------")
      print(f"Generation {i}")
      print(population)
      print(f"Best solution = {population[0]}")
    
    #return population, i
```
